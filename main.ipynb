```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta 

# Gerando histórico de dados dos últimos 30 dias, a cada 15 min:

data_recente = datetime.now() # horário atual
dias_para_subtrair = timedelta(days=29) # timedelta calcula intervalo de tempo (intervalo de tempo de 29 dias, nesse caso)
data_antiga = data_recente - dias_para_subtrair # da data de início, subtrai 6 dias para obter um histórico de 7 dias

datas_e_horarios = pd.date_range(start=data_antiga, end=data_recente, freq='15min') # simula coleta de dados de até 7 dias de 15 em 15 min

# criando DF com coluna DataHora
df = pd.DataFrame({'DataHora': datas_e_horarios})

# extraindo apenas as horas de cada linha da coluna dataHora
df['Hora'] = df['DataHora'].dt.floor('S').dt.time
hora = df['Hora'] # conjunto com todas as horas da coluna Hora do DF 

# extraindo apenas as datas de cada linha da coluna DataHora
df['Data'] = df['DataHora'].dt.date
datas = df['Data'] # conjunto com todas as datas da coluna Data do DF

# inicialização das listas que serão armazenadas no DF
l_irradiacao_solar = []
l_temperatura_painel = []
l_temperatura_ambiente = []
l_status_painel = []

# definindo horários entre 6h e 18h em minutos (horários com incidência solar)
amanhecer = 6
anoitecer = 18

for h in hora:
    if h.hour < anoitecer and h.hour >= amanhecer:
        irradiacao = np.random.randint(100, 1000)
    else:
        irradiacao = 0
    l_irradiacao_solar.append(irradiacao)

    if irradiacao < 0 or irradiacao > 1000:
        raise ValueError("Erro")

    if irradiacao > 0:
        l_temperatura_painel.append(np.random.randint(40, 70))
        l_temperatura_ambiente.append(np.random.randint(24, 35))
        l_status_painel.append(np.random.choice(["Operando...", "Falha"], p=[0.9, 0.1])) # 90% de chance de operar corretamente e 10% de chance de falha
    else:
        l_temperatura_painel.append(np.random.randint(20, 30))
        l_temperatura_ambiente.append(np.random.randint(13, 24))
        l_status_painel.append("Em espera...")

# Removendo a coluna DataHora do DF
df.drop('DataHora', axis=1, inplace=True)

# adicionando demais colunas no DF
df['IrradiacaoSolar (W/m²)'] = l_irradiacao_solar
df['TempPainel'] = l_temperatura_painel
df['TempAmbiente'] = l_temperatura_ambiente
df['StatusPainel'] = l_status_painel

df.head(8)


```
